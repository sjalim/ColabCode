```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import importlib
import nlpaug.augmenter.word as naw
```

```python
aug1 = naw.ContextualWordEmbsAug(model_path='csebuetnlp/banglabert', aug_p=0.1)
```

```python
text = "তুমি একটা বোকা তোমার মত মহিলাদেরকে রান্নাঘরে থাকা উচিত"
print(aug1.augment(text))
```
